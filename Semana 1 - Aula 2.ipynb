#### PROBLEMA

Nome das pessoas está no formato "Sobrenome, Nome", logo, nosso padrão de separação com python está falho.

Vamos usar algo mais potente para pegar esse texto.

Para facilitar nossa vida vamos trocar a vírgula por um "ponto e virgula".

```python
import re
match = re.compile(r'"(.*)(,)(\s.*)"')
match.sub(r'"\1;\3"', linhas[1])
```

Agora que validamos que funciona, vamos criar uma função com esse codigo e extrair nossos dados do arquivo.

```python
def tratar_nome(linha):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"')
    return match.sub(r'"\1;\3"', linha)

tratar_nome(linhas[1])
```

```python
linha = tratar_nome(linhas[1])
dados = linha.split(",")
print(colunas)
print(dados)
```

```python
print(len(colunas), len(dados))
```

#### Ops!

Os indices estão se repetindo, pois temos valores repetidos dentro da lista de dados de uma linha. Vamos ter que fazer diferente.

```python
titanic_dados = {}
for coluna in colunas:
    titanic_dados[coluna] = []

dados_por_linha = []
for linha in linhas[1:]:
    linha = tratar_nome(linha)
    dados_por_linha.append(linha.split(","))
    

for dados in dados_por_linha:
    for indice, dado in enumerate(dados):
        coluna = colunas[indice]
        titanic_dados[coluna].append(dado)
```
