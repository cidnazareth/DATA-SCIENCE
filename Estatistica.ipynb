```python
def simetria(serie):
    media = serie.mean()
    mediana = serie.median()
    moda = serie.mode()[0]
    if (media < mediana < moda):
        print("Negativa")
    elif  (media == mediana == moda):
        print("Normal")
    else #(media > mediana > moda):
        print("Positiva")    

simetria(titanic.Fare)       
```

```python
jogadas = pandas.read_csv("http://bit.ly/jogadas-bc")
jogadas[:3]
```

```python
def variancia(lista):
    n = len(lista)
    mu = np.mean(lista)
    numerador = 0
    for i in lista:
        numerador += (i - mu) ** 2
    return numerador/n

variancia([10,9,11,12,8])
```

```python
amostra = jogadas.sample(frac=0.33) # pega uma amostra (aleatória) de 33% dos meu dados
sns.distplot(amostra, hist=False)
sns.distplot(jogadas.exp, hist=False)
```

```python
print(jogadas.exp.mean())
print(amostra.mean())

print(jogadas.exp.std())
print(amostra.std())
```

```python
amostra_titanic = titanic.Age.dropna()
amostra_titanic.sample(frac=0.33)
sns.distplot(amostra_titanic, hist=False)
```

```python
#intervalo de confiança
from scipy import stats

mu = np.mean(jogadas.exp)
sigma = np.std(jogadas.exp)

stats.norm.interval(0.95, loc=mu, scale=sigma)
```
