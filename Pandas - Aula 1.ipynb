### 4 - Filtrando o dataframe

````python
# Filtrando as linhas onde Pclass == 1
# titanic[ titanic.Pclass == 1 ].head()
# ou
titanic[ titanic["Pclass"] == 1 ].head()
````

O resultado do filtro é outro dataframe, então podemos atribuir o resultado do filtro em uma variavel e trabalhar com ela sem medo de alterar o dataframe original.

````python
sobreviventes = titanic[titanic["Survived"] == 1]
type(sobreviventes)
````

__Filtrando com mais de uma condição__

````python
#  Pessoas que são da primeira classe e sobreviveram

# AND
# Filtrando as linhas onde Pclass == 1 E Survived == 1


filtro_1 = titanic["Pclass"] == 1
filtro_2 = titanic["Survived"] == 1

titanic[ filtro_1 & filtro_2 ].head()

# O mesmo filtro, em uma linha apenas
# titanic[ (titanic["Pclass"] == 1) & (titanic["Survived"] == 1) ].head()
````

````python
#  Pessoas que são da primeira ou segunda classe


# OR
# Filtrando as linhas onde Pclass==1 OU Pclass == 2

titanic[(titanic["Pclass"] == 1) | (titanic["Pclass"] == 2)].head()
````

````python
# Informações estatísticas do Dataframe com o filtro
titanic[(titanic["Pclass"] == 1) & 
        (titanic["Survived"] == 1) ].describe()
````

````python
# Comparando a diferença entre a Idade antes e depois do filtro
media = titanic.Age.mean()
media_filtro = titanic[(titanic["Pclass"] == 1) & 
                       (titanic["Survived"] == 1) ].Age.mean()
print("""
Média de idade: {} \nMédia de 
idade dos sobreviventes da 
primeira classe: {}""".format(media,media_filtro))
````

#### E se quisermos um filtro que retorne os registros com os passageiros da primeira classe que sobreviveram e os passageiros da terceira classe que não sobreviveram?

### 6 - transformação dos dados

Supondo que queremos criar uma coluna onde o valor é a soma entre o `sibsp` e `parch`

````python
# Definindo a função
def soma_sibsp_parch(linha):
    return linha["SibSp"] + linha["Parch"]
````

````python
# Aplicando a função linha a linha do DataFrame
resultado = titanic.apply(soma_sibsp_parch, axis=1)
type(resultado)
````

Obs.: 

`axis = 1` representa que a função é aplicada linha a linha. 

`axis = 0`, que é o valor padrão, representa qua a função é aplicada coluna a coluna.

__Criando uma nova coluna__ 

````python
titanic["relatives"] = resultado
````
